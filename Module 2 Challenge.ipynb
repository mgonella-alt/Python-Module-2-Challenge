```python
# Import the libarary
import pandas as pd

# Import the CSV files
oil = pd.read_csv('oil_price.csv')
gold = pd.read_csv('gold_stocks_price.csv')

# View the oil DataFrame
print(oil.shape)
print(oil.dtypes)
print(oil.head())
print(oil.tail())
```

```python
# View the gold DataFrame
print(gold.shape)
print(gold.dtypes)
print(gold.head())
print(gold.tail())
```

```python
oil["Date"] = pd.to_datetime(oil["Date"], errors="coerce")
gold["Date"] = pd.to_datetime(gold["Date"], errors="coerce")
```

```python
# Gold subset (Date, Open, High, Low)
```

```python
gold_subset = gold[["Date", "Open", "High", "Low"]].iloc[:500].copy()

print("gold_subset shape:", gold_subset.shape)
print(gold_subset.dtypes)
display(gold_subset.head())
display(gold_subset.tail())
```

```python
# Missing Values
```

```python
print("Missing values (gold_subset):")
print(gold_subset.isna().sum())
```

```python
# Min/Max for Open/High/Low
```

```python
print("Min/Max (gold_subset):")
display(gold_subset[["Open", "High", "Low"]].agg(["min", "max"]))
```

```python
#  3) Oil subset (Date, Open, High, Low)
```

```python
oil_subset = oil[["Date", "Open", "High", "Low"]].iloc[:500].copy()

print("oil_subset shape:", oil_subset.shape)
print(oil_subset.dtypes)
display(oil_subset.head())
display(oil_subset.tail())
```

```python
# Missing Values
```

```python
print("Missing values (oil_subset):")
print(oil_subset.isna().sum())
```

```python
# Min/Max for Open/High/Low
```

```python
print("Min/Max (oil_subset):")
display(oil_subset[["Open", "High", "Low"]].agg(["min", "max"]))
```

```python
# 4) Statistical analysis (mean + 75th percentile via describe())
```

```python
gold_stats = gold_subset[["Open", "High", "Low"]].describe(percentiles=[0.75])
oil_stats  = oil_subset[["Open", "High", "Low"]].describe(percentiles=[0.75])

display(gold_stats)
display(oil_stats)

# If you ONLY want mean + 75th percentile:
display(gold_stats.loc[["mean", "75%"]])
display(oil_stats.loc[["mean", "75%"]])
```


Missing values: both gold_subset and oil_subset have 0 missing values in Date/Open/High/Low (in the slices shown).

Gold (Open/High/Low): higher overall levels (mean Open ≈ 149; 75th percentile Open ≈ 161.86). The min/max range suggests noticeable movement across the period (Open min ≈ 117.58, max ≈ 173.20).

Oil (Open/High/Low): lower levels (mean Open ≈ 96.20; 75th percentile Open ≈ 102.00). Range is smaller in absolute terms (Open min ≈ 78.11, max ≈ 110.28).

Interpretation: in this window, gold trades at a higher price level and shows a wider absolute range; oil is lower-priced with a tighter absolute range. Next step would be to plot both over time to compare volatility properly (percent changes / returns are best for that).
